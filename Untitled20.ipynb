## System Health Monitoring with Machine Learning

This notebook explores various machine learning models (Decision Trees and Random Forests) for monitoring system health. It covers data generation, regression, feature importance, scenario analysis, and classification tasks.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Added seaborn import

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, accuracy_score, ConfusionMatrixDisplay

rng = np.random.default_rng(5)
n = 500 # Changed to 500 rows as requested

coolant_flow = rng.uniform(0.7, 5.5, n)
heat_load = rng.uniform(35, 95, n)
ambient_temp = rng.uniform(20, 38, n)
vibration = rng.gamma(shape=2.0, scale=0.35, size=n)
pressure = rng.uniform(50, 150, n) # Added new feature: pressure

thermal_resistance = 1.25 / np.maximum(coolant_flow, 0.35)
low_flow_penalty = np.where(coolant_flow < 2.7, 12.0 + 4.0 * (2.7 - coolant_flow), 0.0)
T_max = (
    42
    + thermal_resistance * heat_load
    + 0.85 * (ambient_temp - 25)
    + 2.2 * vibration
    + low_flow_penalty
    + 0.1 * (pressure - 100) # Incorporating pressure logically
    + rng.normal(0, 2.6, n)
)

overheat = (T_max > 85).astype(int)

X = pd.DataFrame({
    "coolant_flow_L_min": coolant_flow,
    "heat_load_W": heat_load,
    "ambient_temp_C": ambient_temp,
    "vibration_index": vibration,
    "pressure_kPa": pressure, # Added new feature to DataFrame
})
y_reg = T_max
y_cls = overheat

X_train, X_test, y_train, y_test, yc_train, yc_test = train_test_split(
    X, y_reg, y_cls, test_size=0.25, random_state=2, stratify=y_cls
)

models = {
    "shallow_tree_depth_3": DecisionTreeRegressor(max_depth=3, min_samples_leaf=8, random_state=0),
    "deep_tree_unlimited": DecisionTreeRegressor(max_depth=None, min_samples_leaf=1, random_state=0),
    "random_forest": RandomForestRegressor(
        n_estimators=300,
        max_depth=7,
        min_samples_leaf=3,
        max_features="sqrt",
        random_state=0,
    ),
}

print("Regression test RMSE")
for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, pred)) # Removed 'squared=False' and added np.sqrt
    print(f"{name:24s}: {rmse:6.2f} deg C")

fig, ax = plt.subplots(figsize=(12, 5))
plot_tree(
    models["shallow_tree_depth_3"],
    feature_names=X.columns, # Updated feature_names
    filled=True,
    rounded=True,
    fontsize=8,
    ax=ax,
)
ax.set_title("Shallow regression tree: learned feature hierarchy")
plt.show()

importance = pd.DataFrame({
    "feature": X.columns,
    "shallow_tree": models["shallow_tree_depth_3"].feature_importances_,
    "random_forest": models["random_forest"].feature_importances_,
}).sort_values("random_forest", ascending=False)
print(importance.to_string(index=False))

flow_grid = np.linspace(0.7, 5.5, 220)
scenario = pd.DataFrame({
    "coolant_flow_L_min": flow_grid,
    "heat_load_W": 72.0,
    "ambient_temp_C": 30.0,
    "vibration_index": 0.8,
    "pressure_kPa": 100.0, # Added new feature to scenario
})

plt.figure(figsize=(8, 4.6))
for name, model in models.items():
    plt.plot(flow_grid, model.predict(scenario), label=name)
plt.axvline(2.7, color="black", linestyle="--", linewidth=1, label="true low-flow threshold")
plt.xlabel("Coolant flow rate (L/min)")
plt.ylabel("Predicted maximum temperature (deg C)")
plt.title("Effect of max_depth and Random Forest averaging")
plt.grid(alpha=0.25)
plt.legend()
plt.show()

classifier = DecisionTreeClassifier(max_depth=3, min_samples_leaf=8, random_state=0)
classifier.fit(X_train, yc_train)
yc_pred = classifier.predict(X_test)
print("Classification accuracy:", round(accuracy_score(yc_test, yc_pred), 3))

ConfusionMatrixDisplay.from_predictions(
    yc_test,
    yc_pred,
    display_labels=["safe", "overheat"],
    cmap="Blues",
)
plt.title("Decision-tree classifier: thermal safety")
plt.show()

# Add visualization for the classification tree
fig_cls, ax_cls = plt.subplots(figsize=(12, 5))
plot_tree(
    classifier,
    feature_names=X.columns, # Updated feature_names
    class_names=["safe", "overheat"],
    filled=True,
    rounded=True,
    fontsize=8,
    ax=ax_cls,
)
ax_cls.set_title("Decision-tree classifier: learned feature hierarchy")
plt.show()
```

### 1. Data Generation

We start by generating synthetic data that simulates a system's operational parameters and its maximum temperature. The parameters include coolant flow, heat load, ambient temperature, vibration, and **pressure**. A 'low-flow penalty' is introduced to simulate increased temperature under suboptimal coolant flow conditions. An 'overheat' flag is also generated based on a temperature threshold.

```python
rng = np.random.default_rng(5)
n = 500 # Changed to 500 rows as requested

coolant_flow = rng.uniform(0.7, 5.5, n)
heat_load = rng.uniform(35, 95, n)
ambient_temp = rng.uniform(20, 38, n)
vibration = rng.gamma(shape=2.0, scale=0.35, size=n)
pressure = rng.uniform(50, 150, n) # Added new feature: pressure
```

# This section defines the relationships between the input parameters and the system's maximum temperature (`T_max`), including a 'low-flow penalty' and the influence of pressure. An 'overheat' target (`overheat`) is also derived for classification tasks.

```python
thermal_resistance = 1.25 / np.maximum(coolant_flow, 0.35)
low_flow_penalty = np.where(coolant_flow < 2.7, 12.0 + 4.0 * (2.7 - coolant_flow), 0.0)
T_max = (
    42
    + thermal_resistance * heat_load
    + 0.85 * (ambient_temp - 25)
    + 2.2 * vibration
    + low_flow_penalty
    + 0.1 * (pressure - 100)
    + rng.normal(0, 2.6, n)
)

overheat = (T_max > 85).astype(int)
```

# The generated features are compiled into a Pandas DataFrame `X`, and the target variables (`T_max` for regression, `overheat` for classification) are defined. The data is then split into training and testing sets, ensuring a stratified split for the classification target.

```python
X = pd.DataFrame({
    "coolant_flow_L_min": coolant_flow,
    "heat_load_W": heat_load,
    "ambient_temp_C": ambient_temp,
    "vibration_index": vibration,
    "pressure_kPa": pressure,
})
y_reg = T_max
y_cls = overheat

X_train, X_test, y_train, y_test, yc_train, yc_test = train_test_split(
    X, y_reg, y_cls, test_size=0.25, random_state=2, stratify=y_cls
)
```

### 2. Regression Models and Evaluation

Here, we train and evaluate three regression models to predict `T_max`:

*   **Shallow Decision Tree**: A `DecisionTreeRegressor` with `max_depth=3` to capture simple relationships.
*   **Deep Decision Tree**: A `DecisionTreeRegressor` with unlimited depth to see how complex a single tree can get.
*   **Random Forest**: An `RandomForestRegressor` which is an ensemble of multiple decision trees, generally providing better generalization and robustness.

The performance is evaluated using Root Mean Squared Error (RMSE) on the test set.

```python
models = {
    "shallow_tree_depth_3": DecisionTreeRegressor(max_depth=3, min_samples_leaf=8, random_state=0),
    "deep_tree_unlimited": DecisionTreeRegressor(max_depth=None, min_samples_leaf=1, random_state=0),
    "random_forest": RandomForestRegressor(
        n_estimators=300,
        max_depth=7,
        min_samples_leaf=3,
        max_features="sqrt",
        random_state=0,
    ),
}
```

```python
print("Regression test RMSE")
for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    print(f"{name:24s}: {rmse:6.2f} deg C")
```

### 3. Feature Importance and Tree Visualization

This section visualizes the structure of the shallow decision tree and calculates the feature importance for both the shallow tree and the Random Forest model. Feature importance helps understand which input variables contribute most to the model's predictions.

```python
fig, ax = plt.subplots(figsize=(12, 5))
plot_tree(
    models["shallow_tree_depth_3"],
    feature_names=X.columns,
    filled=True,
    rounded=True,
    fontsize=8,
    ax=ax,
)
ax.set_title("Shallow regression tree: learned feature hierarchy")
plt.show()
```

```python
importance = pd.DataFrame({
    "feature": X.columns,
    "shallow_tree": models["shallow_tree_depth_3"].feature_importances_,
    "random_forest": models["random_forest"].feature_importances_,
}).sort_values("random_forest", ascending=False)
print(importance.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot for Shallow Tree Feature Importance
sns.barplot(x='shallow_tree', y='feature', data=importance.sort_values('shallow_tree', ascending=False), ax=axes[0])
axes[0].set_title('Shallow Tree Feature Importance')
axes[0].set_xlabel('Importance')
axes[0].set_ylabel('Feature')

# Plot for Random Forest Feature Importance
sns.barplot(x='random_forest', y='feature', data=importance, ax=axes[1])
axes[1].set_title('Random Forest Feature Importance')
axes[1].set_xlabel('Importance')
axes[1].set_ylabel('Feature')

plt.tight_layout()
plt.show()
```

### 4. Scenario Analysis: Effect of Coolant Flow

We create a hypothetical scenario where `coolant_flow` varies while other parameters are held constant. This allows us to observe how each trained regression model predicts `T_max` as the coolant flow rate changes, and how they respond to the 'true low-flow threshold' defined in the data generation.

```python
flow_grid = np.linspace(0.7, 5.5, 220)
scenario = pd.DataFrame({
    "coolant_flow_L_min": flow_grid,
    "heat_load_W": 72.0,
    "ambient_temp_C": 30.0,
    "vibration_index": 0.8,
    "pressure_kPa": 100.0,
})
```

```python
plt.figure(figsize=(8, 4.6))
for name, model in models.items():
    plt.plot(flow_grid, model.predict(scenario), label=name)
plt.axvline(2.7, color="black", linestyle="--", linewidth=1, label="true low-flow threshold")
plt.xlabel("Coolant flow rate (L/min)")
plt.ylabel("Predicted maximum temperature (deg C)")
plt.title("Effect of max_depth and Random Forest averaging")
plt.grid(alpha=0.25)
plt.legend()
plt.show()
```

### 5. Classification Model for Thermal Safety

Finally, we train a `DecisionTreeClassifier` to predict the `overheat` condition. This is a binary classification task to determine if the system is likely to overheat. The model's performance is evaluated using accuracy and a confusion matrix to understand its true positive, true negative, false positive, and false negative rates.

```python
classifier = DecisionTreeClassifier(max_depth=3, min_samples_leaf=8, random_state=0)
classifier.fit(X_train, yc_train)
yc_pred = classifier.predict(X_test)
print("Classification accuracy:", round(accuracy_score(yc_test, yc_pred), 3))
```

```python
ConfusionMatrixDisplay.from_predictions(
    yc_test,
    yc_pred,
    display_labels=["safe", "overheat"],
    cmap="Blues",
)
plt.title("Decision-tree classifier: thermal safety")
plt.show()
```

```python
fig_cls, ax_cls = plt.subplots(figsize=(12, 5))
plot_tree(
    classifier,
    feature_names=X.columns,
    class_names=["safe", "overheat"],
    filled=True,
    rounded=True,
    fontsize=8,
    ax=ax_cls,
)
ax_cls.set_title("Decision-tree classifier: learned feature hierarchy")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot for Shallow Tree Feature Importance
sns.barplot(x='shallow_tree', y='feature', data=importance.sort_values('shallow_tree', ascending=False), ax=axes[0])
axes[0].set_title('Shallow Tree Feature Importance')
axes[0].set_xlabel('Importance')
axes[0].set_ylabel('Feature')

# Plot for Random Forest Feature Importance
sns.barplot(x='random_forest', y='feature', data=importance, ax=axes[1])
axes[1].set_title('Random Forest Feature Importance')
axes[1].set_xlabel('Importance')
axes[1].set_ylabel('Feature')

plt.tight_layout()
plt.show()
```
